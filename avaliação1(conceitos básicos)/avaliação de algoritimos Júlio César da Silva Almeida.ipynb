```python
valortotal = float(input('Insira o valor da conta a ser paga: '))
if valortotal <= 0:
    print('Valor inválido')
else:
    resto = valortotal % 3
    divisão = int(valortotal / 3)
    valorcarlos = divisão
    valorandre = divisão
    valorfelipe = divisão + resto



print(f" Valor a ser pago por carlos: {valorcarlos} \n Valor a ser pago por andré: {valorandre} \n Valor a ser pago por felipe: {valorfelipe}")





```

```python
valorvenal = float(input("Insira o valor venal do seu carro: "))
if valorvenal <= 0:
    print("Valor invalido")
else:
    tipodeveiculo = input("Qual o seu modelo de carro? (ex: automovél, caminhonete, furgão, taxí, escolar, motocicletas, alugados, ônibus, micro-ônibus, caminhão e caminhão trator: ")
    lista = ["automovél", "caminhonete", "furgão", "taxí", "escolar", "motocicletas", "alugados", "ônibus", "micro-ônibus", "caminhão", "caminhão trator"]

    if tipodeveiculo not in lista:
        print('Véiculo inválido!')
    else:
        if tipodeveiculo == 'automovél':
            ipva = (valorvenal * 0.04)
        elif tipodeveiculo == 'caminhote de carga' or 'furgão':
            ipva = (valorvenal * 0.03)
        elif tipodeveiculo == 'taxí' or 'escolar':
            ipva = (valorvenal * 0.02)
        elif tipodeveiculo == 'motocicleta':
            ipva = (valorvenal * 0.02)
        elif tipodeveiculo == 'alugados':
            ipva = (valorvenal * 0.01)
        elif tipodeveiculo == 'ônibus' or 'micro-ônibus' or 'caminhão' or 'caminhão trator':
            ipva = (valorvenal * 0.01)
        print('O valor a ser pago do seu ipva será:', ipva)
```
